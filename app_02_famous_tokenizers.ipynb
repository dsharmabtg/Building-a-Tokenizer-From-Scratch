# TOKENIZERS EXAMPLES AND EXAMINATION

## BERT

### Bert is an encoder transformer - its tokenizer doesn't have a decode function to show us the output of the model as word because its output is a vector.

```python
from transformers import BertTokenizer, BertModel
bert_tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
text = "joyfully"
encoded_input = bert_tokenizer(text)
print(encoded_input)
```

What are these tokens? lets see one by one:

```python
txt = bert_tokenizer.convert_ids_to_tokens([101])
print(txt)
txt = bert_tokenizer.convert_ids_to_tokens([6569])
print(txt)
txt = bert_tokenizer.convert_ids_to_tokens([7699])
print(txt)
txt = bert_tokenizer.convert_ids_to_tokens([102])
print(txt)
```

## GPT-2

### GPT2 is an autoregressive transformer.  Its output are words. The tokenizer's decode function will take the output of the model and decode it to words.

```python
from transformers import GPT2Tokenizer, GPT2Model
gpt2_tokenizer = GPT2Tokenizer.from_pretrained('gpt2-medium')
text = "joyfully"
encoded_input = gpt2_tokenizer(text)
print(encoded_input)
```

```python
from transformers import GPT2Tokenizer

# Instantiate the tokenizer
gpt2_tokenizer = GPT2Tokenizer.from_pretrained('gpt2-medium')

# Convert token ids to tokens
txt = gpt2_tokenizer.convert_ids_to_tokens(ids=[2633, 2759])  # ids should be a list
print(txt)
```
